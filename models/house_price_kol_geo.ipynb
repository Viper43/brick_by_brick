```python
def remove_ppua_outliers(df):
    df_out = pd.DataFrame()

    for key, subdf in df.groupby('CITY'):
        mean = subdf.PRICE_SQFT.mean()
        sd = subdf.PRICE_SQFT.std()

        # Filter out outliers based on mean and standard deviation

        reduced_df = subdf[((subdf.PRICE_SQFT >= (mean - sd)) & (subdf.PRICE_SQFT <= (mean + 1.3*sd)))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    
    return df_out
```

```python
df6 = remove_ppua_outliers(df5)
df6.shape
```

```python
df6.BEDROOM_NUM.unique()
```

```python
def plot_scatter_chart(df, city, xlim, ylim):
    bhk2 = df[(df.BEDROOM_NUM == 2) & (df.CITY == city)]
    bhk3 = df[(df.BEDROOM_NUM == 3) & (df.CITY == city)]
    bhk4 = df[(df.BEDROOM_NUM == 4) & (df.CITY == city)]

    matplotlib.rcParams['figure.figsize'] = (15, 10)
    plt.scatter(bhk2.AREA, bhk2.TOTAL_PRICE, label='2 BHK', color='blue', s=50)
    plt.scatter(bhk3.AREA, bhk3.TOTAL_PRICE, label='3 BHK', marker='+', color='green', s=50)
    plt.scatter(bhk4.AREA, bhk4.TOTAL_PRICE, label='4 BHK', marker='*', color='red', s=50)

    plt.xlim(0, xlim)
    plt.ylim(0, ylim)
    plt.xlabel("Area (in sqft)")
    plt.ylabel("Total Price (in lakhs)")
    plt.title(f"Price vs Area for {city}")
    plt.legend()
    plt.show()
```

```python
plot_scatter_chart(df6, 'Kolkata South', 3000, 500)
plot_scatter_chart(df6, 'Kolkata North', 2500, 300)
plot_scatter_chart(df6, 'Kolkata East', 2500, 300)
plot_scatter_chart(df6, 'Kolkata West', 1500, 100)
plot_scatter_chart(df6, 'Kolkata Central', 2500, 300)
```

```python
df7 = df6.drop(columns=['PRICE_SQFT'])
df7.head()
```

```python
dummies = pd.get_dummies(df7.CITY, dtype='int')
dummies.groupby('Kolkata North')['Kolkata North'].agg('count')
```

```python
df8 = pd.concat([df7, dummies.drop('Kolkata Central', axis=1)], axis=1)
df8.head()
```

```python
df9 = df8.drop(columns=['CITY'])
df9.shape
```

```python
x = df9.drop(['TOTAL_PRICE'], axis=1)
y = df9['TOTAL_PRICE']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(linear_model.LinearRegression(), x, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
```

```python
def get_best_model(x, y):
    algos = {
        'linear_regression': {
            'model': linear_model.LinearRegression(),
            'params': {}
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        'ridge': {
            'model': Ridge(),
            'params': {
                'alpha': [1, 2],
                'solver': ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['squared_error', 'friedman_mse'],
                'splitter': ['best', 'random'],
            }
        }

    }


    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv= cv, return_train_score=False)
        gs.fit(x, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
            
        })
    #print(gs.cv_results_)
    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
    # return pd.DataFrame(gs.cv_results_)

get_best_model(x, y)
```

```python
# selecting and training the best model
best_model = DecisionTreeRegressor(criterion='friedman_mse', splitter='random', random_state=83)
best_model.fit(x_train, y_train)
best_model.score(x_test, y_test)
```
